```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./rsc/spotify2023.csv",encoding='ISO-8859-1')
```

```python
df.head()
```

```python
df.columns = ['track_name', 'artist_name', 'artist_count', 'released_year','released_month', 'released_day', 'in_spotify_playlists','in_spotify_charts', 'streams', 'in_apple_playlists', 'in_apple_charts','in_deezer_playlists', 'in_deezer_charts', 'in_shazam_charts', 'bpm','key', 'mode', 'danceability', 'valence', 'energy',  'acousticness', 'instrumentalness', 'liveness', 'speechiness']
```

```python
#finding null values in each column
print(df.isnull().sum())

# change in_shazam_charts null to 0 and key null to C
df['in_shazam_charts'] = df['in_shazam_charts'].fillna(0)
df['key'] = df['key'].fillna('C')

```

```python
df.loc[df["streams"] == "218508051"]
```

```python
df.loc[df["streams"]== "BPM110KeyAModeMajorDanceability53Valence75Energy69Acousticness7Instrumentalness0Liveness17Speechiness3", "streams"] = 218508051
```

```python
df["streams"] = df["streams"].astype(np.int64)
```

```python
# Replace all "," with "" in the in_deezer_playlists column and change the column to in
df["in_deezer_playlists"] = df["in_deezer_playlists"].str.replace(',', '').astype(int)
```

```python
# convert object having , in in_shazam_charts to int with no ,
df["in_shazam_charts"] = df["in_shazam_charts"].str.replace(',', '').astype(float)
```

```python
print(df["in_shazam_charts"].value_counts())
#show data with columns track_name and in_shazam_charts only having null in in_shazam_charts
print(df[df["in_shazam_charts"].isnull()][["track_name","in_shazam_charts"]])
#convert NaN to 0
df["in_shazam_charts"] = df["in_shazam_charts"].fillna(0)
```

```python
# change in_shazam_charts float to int
df["in_shazam_charts"] = df["in_shazam_charts"].astype(int)
```
